Importing Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
```

Reading the dataset

```python
data=pd.read_csv("HRDataset_v14.csv")
```

```python
data.head()
```

Exploring the dataset

```python
data.describe()
```

```python
data.isnull().sum()
```

Plotting Graph

```python
gender_employment_source_counts = data.groupby(['Sex', 'EmploymentStatus', 'RecruitmentSource']).size().unstack()


fig, fig = plt.subplots(figsize=(10, 6))
gender_employment_source_counts.plot(kind='bar', stacked=True, ax=fig)

fig.set_xlabel('Gender and Employment Status')
fig.set_ylabel('Count')
fig.set_title('Distribution of Employment Status, Recruitment Source by Gender')
fig.legend(bbox_to_anchor=(1.05, 1))

plt.tight_layout()
```
